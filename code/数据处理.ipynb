# 待续

```python
# 打开文件，使用"r"参数表示读取模式（read mode）
src_path = "../output/one_times/"
res_path = "../output/one_times/"
for i in range(3):
    print(i)
    with open(src_path + str(i) + ".txt", 'r', encoding="utf8") as file:
        file_content = file.readlines()
    q = []
    a = []
    for idx, name in enumerate(file_content):
        if idx == 1998:
            q.append(name.split("\t")[0])
            a.append(name.split("\t")[1][0])
        elif idx != 1998:
            if idx % 2 == 0:
                q.append(name.split("\n")[0])
            elif idx % 2 != 0:
                pattern = r'\b\d\b'
                data = re.search(pattern, name)[0]
                a.append(data)
    filedata = pd.DataFrame([q, a]).T
    print(pd.DataFrame(filedata)[1].value_counts())
    filedata.to_csv(res_path + str(i) + ".txt", sep="\t", index=False, header = None)
```

# 根据每个类别的文本分词组成新的样本

```python
data0 = []
data1 = []
data2 = []
data3 = []
data4 = []
data5 = []
for i in range(7000):
    if df3[1][i] == 0:
        data0.append(df3[0][i])
    elif df3[1][i] == 1:
        data1.append(df3[0][i])
    elif df3[1][i] == 2:
        data2.append(df3[0][i])
    elif df3[1][i] == 3:
        data3.append(df3[0][i])
    elif df3[1][i] == 4:
        data4.append(df3[0][i])
    elif df3[1][i] == 5:
        data5.append(df3[0][i])
```

0 诊断
1 治疗     : 和其他病或者症状的关系
2 医疗常识
3 健康生活方式 
4 流行病学
5 其他

```python
import jieba
all_text = []
for i in [data0, data1, data2, data3, data4, data5]:
    text = []
    for j in range(len(i)):
        seg_list = jieba.lcut(i[j], cut_all=False)  # cut_all=False 表示精确模式
        text.append(seg_list)
    all_text.append(text)
all_text[2]
```

```python
all_cat_text = []
for i in range(6):
    df_yhl = all_text[i]
    all_data = []
    for j in range(300):
        text = []
        random_integers = [random.randint(1, pd.DataFrame(df_yhl).shape[0]) for _ in range(5)]
        for idx, n in enumerate(random_integers):
            df_text = df_yhl[n-1]
            if idx == 0:
                text.append(df_text[0])
            else:
                random_integer = random.randint(1, pd.DataFrame(df_text).shape[1])
                text.append(df_text[random_integer])
        text = "".join(text)
        all_data.append(text)
    all_cat_text.append(all_data)
```
